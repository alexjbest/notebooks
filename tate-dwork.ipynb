Tate-Dwork
----------------

Experiments with the math in Tate's letter to Dwork.

First we choose a curve and a prime with lots of $\mathbf F_p$-rational $p$-torsion, lets let

```python
p = 3
```

https://www.lmfdb.org/EllipticCurve/6.6.722000.1/76.1/b/1

```python
R.<x> = PolynomialRing(QQ);  K.<a> = NumberField(R([1, -5, 4, 7, -6, -1, 1]), embedding=1)
E = EllipticCurve([K([4,-10,-8,12,1,-2]),K([5,-10,-8,12,1,-2]),K([1,0,0,0,0,0]),K([-5,10,8,-12,-1,2]),K([0,0,0,0,0,0])])
E = E.short_weierstrass_model(complete_cube=False)
```

```python
k3 = K.prime_above(3).residue_field()
```

```python
Ek3 = E.change_ring(k3)
Ek3.cardinality().factor()
```

```python
P = E.torsion_subgroup().gen(0); P
```

```python
Ek3(P).order()
```

```python
Ehat = E.formal_group(); Ehat
```

```python
(Ehat.x().derivative()/(2*Ehat.y())).integral()
```
