# Param points
a la Cohen

We are looking for triples $x_1,x_2,x_3$ in arith prog
$$
y^2 = x^3 - D^2 x
$$
with $x_i = k_i^2$ so that
$$
(\frac y k)^2 = k^4 - D^2
$$
thus given a solution $a,b,c$ of
$$a^2 + b^2 = c^4$$
we can set
$k = c$
and either
$D = b$,
$y = ak = ac$ or 
$D = a$,
$y = bk = bc$.
According to Cohen we have
$$(a,b,c) = \left(\pm\left(s^{2}-2 s t-t^{2}\right)\left(s^{2}+2 s t-t^{2}\right), 4 t s(s-t)(s+t), \pm\left(s^{2}+t^{2}\right)\right)$$
or likewise with $a,b$ swapped.

```python
def abc_from_st(s,t,sign=1):
    return (sign*(s^2-2*s*t-t^2)*(s^2+2*s*t-t^2), 4*t*s*(s-t)*(s+t),sign*(s^2+t^2))

a,b,c = (abc_from_st(1,2,sign=-1))
print(a,b,c)
print(a^2 + b^2 == c^4)
def kyD_from_st(s,t,sign=1,swap=False): # TODO swaps
    a,b,c = (abc_from_st(s,t,sign=sign))
    if swap:
        a,b = b,a
    return (c,a*c,b)
k,y,D= (kyD_from_st(1,2,sign=1))
print(k,y,D)
print((y/k)^2 == k^4 - D^2)
print(y^2 == k^6 - D^2*k^2)
```
